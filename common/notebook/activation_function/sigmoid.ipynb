# シグモイド関数（Sigmoid Function）

$$
sigmoid(x) = \frac{1}{1 + exp(-x)}
$$

```python
import numpy as np

def sigmoid(x):
    """Sigmoid function.
    
    x: input data
    """
    return 1 / (1 + np.exp(-x))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.title('Sigmoid function')
plt.ylim(-0.1, 1.1)
plt.show()
```
