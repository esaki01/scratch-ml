# ReLU関数（ReLU Function）

$$
relu(x) = max(0, x)
$$

```python
import numpy as np

def relu(x):
    """ReLU function.
    
    x: input data
    """
    return np.maximum(0, x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.title('ReLU function')
plt.ylim(-1, 5)
plt.show()
```
